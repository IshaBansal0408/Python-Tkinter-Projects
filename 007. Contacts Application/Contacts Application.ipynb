```python
def show_all():
    t.configure(state='normal')
    t.delete('1.0', END)
    f = open('contacts.txt','r')
    x = f.readlines()
    for i in x:
        t1=i.split(',')
        s=str(t1[0])+" : "+str(t1[1].split('\n')[0])+'\n'
        t.insert(END,s)
    t.configure(state='disabled')
    f.close()

def add_contact():
    c=e1.get()
    n=e2.get()
    f = open('contacts.txt','a')
    if(len(n)==10):
        f.write(c+","+str(n)+"\n")
        f.close()
        messagebox.showinfo('Info',"Success!")
    else:
        exitmsg=messagebox.showerror("Error!","Not Valid Number!")
    f.close()
    show_all()

def clear_list():
    f = open('contacts.txt','w')
    show_all()
    f.close()

def exit():
    exitmsg=messagebox.askyesno("EXIT TO-DO List?","Please Confirm!")
    if exitmsg>0:
        root.destroy()
        return

def delete_contact():
    g=e3.get()
    f = open('contacts.txt','r')
    x = f.readlines()
    f = open('contacts.txt','w')
    for i in x:
        z=i.split(",")
        if(z[0]!=g):
            h=z[0]+","+z[1].split('\n')[0]
            f.write(h+"\n")
    f.close()
    show_all()
```

```python
from tkinter import*
from tkinter import messagebox
root=Tk()
root.title("Contacts")
root.geometry("400x800")
root.resizable(width=False,height=False)
root.configure(bg="#cc3300")

l1=Label(root,text="Manage Contacts",font=("Times New Roman",30,"bold"),pady=10,bg="#cc3300",fg="White")
l1.pack()

v1=StringVar()
l2=Label(root,text="Enter Contact Name:",font=("Times New Roman",15,"bold"),pady=10,bg="#cc3300",fg="Black")
e1=Entry(root,font=("Times New Roman",20,"bold"),textvariable=v1)
l2.pack()
e1.pack(pady=5)

v2=IntVar()
l3=Label(root,text="Enter Contact Number:",font=("Times New Roman",15,"bold"),pady=10,bg="#cc3300",fg="Black")
e2=Entry(root,font=("Times New Roman",20,"bold"),textvariable=v2)
l3.pack()
e2.pack(pady=5)

b1=Button(root,text="Enter Contact",font=("Times New Roman",15,"bold"),command=add_contact)
b1.pack(pady=10)

b2=Button(root,text="Show List",font=("Times New Roman",15,"bold"),command=show_all)
b2.pack(pady=10)

t=Text(height=15,width=40,state="disabled")
t.pack()

v3=StringVar()
l4=Label(root,text="Enter Name to be deleted:",font=("Times New Roman",15,"bold"),bg="#cc3300",fg="Black")
e3=Entry(root,font=("Times New Roman",20,"bold"),textvariable=v3)
b4=Button(root,text="Delete Contact",font=("Times New Roman",15,"bold"),command=delete_contact)
l4.pack()
e3.pack()
b4.pack(pady=10)

b3=Button(root,text="Clear List",font=("Times New Roman",15,"bold"),command=clear_list)
b3.place(x=70,y=750)
b5=Button(root,text="Exit Application",font=("Times New Roman",15,"bold"),command=exit)
b5.place(x=200,y=750)

root.mainloop()
```
